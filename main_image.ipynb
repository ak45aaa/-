```python
import cv2
from paddleocr import PaddleOCR
```

```python
ocr_kor = PaddleOCR(lang='korean')
ocr = PaddleOCR()
```

```python
image1 = cv2.imread('1_2.jpg')
```

```python
import matplotlib.pyplot as plt

plt.imshow(image1)
plt.axis('off')
plt.show()
```

```python
result = ocr.ocr(image1)
result_kor = ocr_kor.ocr(image1)
```
